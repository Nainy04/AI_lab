```python


import numpy as np 
import pandas as pd 
import os
import cv2
from matplotlib import pyplot as plt



folder_path = '/kaggle/input/deepfake-and-real-images/Dataset/Test/Fake'

image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

num_to_load = 10  
images = []

for file_name in image_files[:num_to_load]:
    img_path = os.path.join(folder_path, file_name)
    img = cv2.imread(img_path)
    
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.axis('off')  
    plt.show()

    print(img_path)
    if img is not None:
        images.append(img)
    else:
        print(f"Failed to load {img_path}")

print(f"Loaded {len(images)} images for preview.")

```

```python
from PIL import Image
images_to_display = image_files[:16]
ROWS = 4
COLS = 4
NUM_IMAGES = ROWS * COLS


fig, axes = plt.subplots(ROWS, COLS, figsize=(10, 10)) 


axes = axes.flatten()

for i, filename in enumerate(images_to_display):
    
    img_path = os.path.join(folder_path, filename)
    
    
    try:
        img = Image.open(img_path)
    except Exception as e:
        print(f"Error loading {filename}: {e}")
        continue
        
    # 3. Display the image on the current subplot
    axes[i].imshow(img)
    
    # 4. Remove axis ticks and labels for a cleaner look
    axes[i].axis('off') 
    
    # Optional: Set the title to the file name (you can use class labels if you have them)
    axes[i].set_title(filename[:10] + '...', fontsize=8) 


plt.tight_layout()

plt.show()
```

```python
import os

image_directory = '/kaggle/input/deepfake-and-real-images/Dataset/Test' 

total_size_bytes = 0

try:
   
    for dirpath, dirnames, filenames in os.walk(image_directory):
        for filename in filenames:
            # Construct the full path to the file
            file_path = os.path.join(dirpath, filename)
            
            # Add the size of the individual file to the total
            # os.path.getsize(file_path) returns the size in bytes
            total_size_bytes += os.path.getsize(file_path)

    # --- Convert to a readable format (Optional, but highly recommended) ---
    
    # Calculate the size in Megabytes (MB)
    size_in_mb_test = total_size_bytes / (1024 * 1024) 

    print(f"Directory: {image_directory}")
    print(f"Total Size of ALL Images and Files: {total_size_bytes} bytes")
    print(f"Total Size: {size_in_mb_test:.2f} MB")

except FileNotFoundError:
    print(f"Error: The directory '{image_directory}' was not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
